```python
import pandas as pd
import numpy as np
from cx_analysis.dataframe_tools import assemble_cxvectors
```

```python
# link_df is a dataframe with an entry for every synapse in the lamina connectome
tp = '210809'
data_path = f"Data/"
#cx = pd.read_pickle(data_path + "_cxdf.pickle")
links = pd.read_pickle(data_path + "210809_linkdf.pickle")
all_cxvecs = assemble_cxvectors(links, excl_unknowns=False)
```

## Table 2 (supplement)
Raw connection counts 

```python
# Table of connections that was observed > 1 in all 29 cartridges

cxtype_sum = all_cxvecs.sum()
display(cxtype_sum)
pub_table = all_cxvecs.T.loc[cxtype_sum.sort_values(ascending=False).index] # sort connection types by sum across om
pub_table = pub_table.loc[[k for k, v in pub_table.sum(axis=1).items() if v > 0]]  # filter out those never observed

unk_rows = pub_table.loc[[k for k in pub_table.index if 'UNKNOWN' in str(k)]]
unk_sum = unk_rows.sum()

pub_table = pub_table.drop(labels=[k for k in pub_table.index if 'UNKNOWN' in str(k)], axis=0)
pub_table.loc["->UNKNOWN"] = unk_sum.T

pub_table

#unknown_rows.sum()
#pub_table.to_excel(f'~/Desktop/lamina_cx-table.xlsx')
```
